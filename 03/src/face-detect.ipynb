# 顔検出

## OpenCVのHaar-like特徴分類器を用いた顔検出を行う
データセット: url(https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_alt.xml)

```python
import matplotlib.pyplot as plt
import cv2
```

```python
# カスケードファイルを指定して検出器を作成
cascade_file = "../dataset/haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)
```

```python
# 画像をグレースケールに変換
img = cv2.imread("../images/girl.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 顔認識を実行
face_list = cascade.detectMultiScale(img_gray, minSize=(150, 150))
if not len(face_list):
    print("failed")
    quit()

for (x, y, w, h) in face_list:
    print("顔の座標 = ", x, y, w, h)
    red = (0, 0, 255)
    cv2.rectangle(img, (x, y), (x+w, y+h), red, thickness=20)
```

```python
cv2.imwrite("../images/face-detect.png", img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
