# アヤメデータの分類

## 目的
1. csvファイルのインポート方法を覚える
2. 簡単な機械学習に触れてみる

## ゴール設定
1. データをスクリプトに取り込む
2. データを数的情報と品種情報に分割する
3. 全データのうち80%を学習用データ, 20%をテストデータに分離する
4. 学習を行い, 正しく分類できているか評価する

## アルゴリズム
SVC - サポートベクター分類(2次元)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#データの読み込み
iris_data = pd.read_csv("csv/iris.csv", encoding="utf-8")

# データをラベルと入力データに分離
y = iris_data.loc[:, "Name"]
x = iris_data.loc[:, ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]]

# 学習用とテスト用に分類
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, train_size = 0.8, shuffle = True)

# 学習
clf = SVC()
clf.fit(x_train, y_train)

# 評価
y_pred = clf.predict(x_test)
print("正解率 = ", accuracy_score(y_test, y_pred))
```

### 補足説明
・sklearn.model_selection.train_test_split() - 配列or行列をランダムなトレーニングとテストのサブセットに分割する

・ sklearn.svm.SVC() - SVCを行う

・pandas.DataFrame.loc[行, 列] - df内で条件を満たした行, 列を抽出する

### 評価
正解率 = 0.96666...から正しく分類できていると評価する
